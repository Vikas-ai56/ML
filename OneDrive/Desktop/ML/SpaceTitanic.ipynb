```python
'''
 Fit: The transformer learns the necessary parameters from the training data
 Transform: The transformer applies the learned parameters to the data to transform it
 Imputer : to replace missing values in a dataset with estimated values by using different strategies

Use fit_transform() only on the training data.
Use transform() on the testing data, to apply the same transformations learned from the training data.
Never: use fit_transform() on the testing data, 
as it would learn new parameters from the test data, leading to data leakage and potentially incorrect results.
'''
```

```python
import sklearn.compose
from sklearn.impute import SimpleImputer
import sklearn.metrics
import sklearn.model_selection
import sklearn.neighbors
from sklearn.pipeline import Pipeline

# Extract PassengerId from test data
test_passenger_ids = test_df['PassengerId']


# Preprocessing for numerical data
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

# Preprocessing for categorical data
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Bundle preprocessing for numerical and categorical data
preprocessor = sklearn.compose.ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])

# Define the model
model = sklearn.neighbors.KNeighborsClassifier()

# Create and evaluate the pipeline
clf = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', model)])

# Split the data into training and validation sets
X_train, X_val, y_train, y_val = sklearn.model_selection.train_test_split(train_df, y, test_size=0.2, random_state=42)

# Hyperparameter tuning using GridSearchCV
param_grid = {
    'classifier__n_neighbors': [3, 5, 7, 9, 11],
    'classifier__weights': ['uniform', 'distance'],
    'classifier__metric': ['euclidean', 'manhattan']
}

grid_search =  sklearn.model_selection.GridSearchCV(clf, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Best parameters and best score
print(f'Best parameters: {grid_search.best_params_}')
print(f'Best cross-validation accuracy: {grid_search.best_score_}')

# Evaluate the model on the validation set
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_val)

accuracy =  sklearn.metrics.accuracy_score(y_val, y_pred)
print(f'Validation Accuracy: {accuracy}')


# Predict on test data
predictions = best_model.predict(test_df)
```
