```python
import tensorflow as tf

# Function definition
def df_to_dataset(dataframe, shuffle=True, batch_size=32):
    original = dataframe.copy()
    labels = dataframe.pop('GradeClass')  
    ds = tf.data.Dataset.from_tensor_slices((dict(original), labels))
    if shuffle:
        ds = ds.shuffle(buffer_size=len(dataframe))
    ds = ds.batch(batch_size)
    ds = ds.prefetch(batch_size) # overlaps data preprocessing and model execution IMPROVES PERFORMANCE 
    return ds


df
```

```python
import sklearn.model_selection
train, test = sklearn.model_selection.train_test_split(df, test_size=0.2)
train, val = sklearn.model_selection.train_test_split(train, test_size=0.2)
train.to_csv('../ml/housing-train.csv', encoding='utf-8', index=False)
test.to_csv('../ml/housing-test.csv', encoding='utf-8', index=False)
val.to_csv('../ml/housing-val.csv', encoding='utf-8', index=False)
```

```python
train_ds = df_to_dataset(train)
val_ds = df_to_dataset(val,shuffle=False, batch_size=32)
train_ds
```

```python
numericColumns = ['Age','Gender','Ethnicity','ParentalEducation','Absences','Tutoring'	
                  'Extracurricular','Sports','Music',]
```

```python
from tensorflow import feature_column as fc

def get_scal(feature):
    def minmax(x):
        mini = train[feature].min()
        maxi = train[feature].max()
        return (x - mini)/(maxi-mini)
    return minmax    

inputs = {
        colname: tf.keras.layers.Input(name=colname, shape=(), dtype='float32')
        for colname in numericColumns
    }

feature_columns = dict()
for header in numericColumns:
    scal_input_fn = get_scal(header)
    feature_columns[header] = fc.numeric_column(header,normalizer_fn=scal_input_fn)
```

```python
feature_layer = tf.keras.layers.Dense(feature_columns, dtype='float64')(inputs)

# `tf.keras.Sequential()` groups a linear stack of layers into a tf.keras.Model.
model = tf.keras.Sequential([
  feature_layer,
  tf.keras.layers.InputLayer(input_shape=feature_columns["Age"].shape()),
  tf.keras.layers.Dense(12, input_dim=8, activation='relu'),
  tf.keras.layers.Dense(8, activation='relu'),
  tf.keras.layers.Dense(1, activation='linear',  name='median_house_value')
])

# Model compile
model.compile(optimizer='adam',
              loss='mse',
              metrics=['mse'])

# Model Fit
history = model.fit(train_ds,
                    validation_data=val_ds,
                    epochs=32)
```
